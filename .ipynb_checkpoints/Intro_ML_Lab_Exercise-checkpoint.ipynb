## Model Training

### Data Preperations

```python
# separate train data into train/val of features and target
X = trainD.drop(columns=['Class'], axis = 1)
Y = trainD['Class']
print(X.shape)
print(Y.shape)
# separate test data into features and target as well
Xtest = testD.drop(columns=['Class'], axis = 1)
Ytest = testD['Class']
print(Xtest.shape)
print(Ytest.shape)
```

```python
# split train data in train and validation
from sklearn.model_selection import train_test_split
Xtrain, Xval, Ytrain, Yval = train_test_split(X, Y, test_size = 0.2, random_state = 0)
print(Xtrain.shape)
print(Ytrain.shape)
print(Xval.shape)
print(Yval.shape)
```

### Model Construction

```python
# import logistic module
from sklearn.linear_model import LogisticRegression
```

```python
# instantiate model
logR_model = LogisticRegression()
# train model by fitting (Xtrain, Ytrain)
logR_model = logR_model.fit(Xtrain, Ytrain)
```

```python
# evaluate logistic mode performance
from sklearn.metrics import accuracy_score
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
from sklearn.metrics import f1_score

from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
```

```python
# make predictions using validation data
val_predictions = logR_model.predict(Xval)
# display all the predictions made
val_predictions
```

```python
print('Validation Scores')
print(f'accuracy_score: {accuracy_score(Yval, val_predictions)}')
print(f'precision_score: {precision_score(Yval, val_predictions)}')
print(f'recall_score: {recall_score(Yval, val_predictions)}')
print(f'f1_score: {f1_score(Yval, val_predictions)}')
```

```python
# validation confusion matrix
target_names = ['1', '2']
conf_mat = confusion_matrix(Yval, val_predictions)
f, ax= plt.subplots(1, 1, figsize=(7,6), dpi=100)
sb.heatmap(conf_mat, annot=True, fmt='.0f', ax = ax, annot_kws={"size":18})
ax.set_xticklabels(target_names,fontsize = 15)
ax.set_yticklabels(target_names,fontsize = 15)
ax.set_xlabel('Predicted labels')
ax.set_ylabel('True labels'); 
ax.set_title('Confusion Matrix Logistic Regression (Validation)');
plt.show()

print('+=+'*18)
print('+=++=++=++=+ Classification Report Project Data'+' '+'+=++=+')
print('+=+'*18)
print(classification_report(Yval, val_predictions,target_names=target_names))
```

- **Confusion Matrix:** a table showing correct predictions and types of incorrect predictions.

- **Precision:** the number of true positives divided by all positive predictions. Precision is also called Positive Predictive Value. It is a measure of a classifier’s exactness. Low precision indicates a high number of false positives.

- **Recall:** the number of true positives divided by the number of positive values in the test data. Recall is also called Sensitivity or the True Positive Rate. It is a measure of a classifier’s completeness. Low recall indicates a high number of false negatives.

- **F1 Score:** the weighted average of precision and recall.

**We can improve the above model by applying the concept of hyperparameter tuning.**

**HYPERPARAMETERS**

Hyperparameters are tuning measures that are specified for a model before training. Each type of model has its own unique set of hyperparameters. 

Oftentimes, the base function has defaults for these hyperparameters when they are not specified. So, it is always important to check the documentation on each model to know the complete list of hyperparameters that can be tuned.

Now that we know the corresponding hyperparameters to tune for our model, the problem now is, how do we combine these hyperparameters to know which combination will give us the best performance for our data?

**RESPONSE**

There is a super easy way to do this using three functions in scikit-learn.

* GridSearchCV

* RandomizedSearchCV

* Pipeline (feeds list of diff models and their hyperparameters to a grid or randomized search)



If not satisfied --> use Cross Validation

```python
from sklearn.model_selection import GridSearchCV, RandomizedSearchCV
```

**GridSearchCV**

```python
# Let's use use gridsearch to finetune our linear regression model
model = LogisticRegression()

# create list of candidate penalty hyperparameters
penalty = ['l1','l2']

# create range of candidate regularization hyperparameters
C = np.logspace(0,4,10)

# create a dictionary hyperparameter candidates
hyperparameters = dict(C=C, penalty=penalty)

# create a grid search object and pass in the hyperparameters defined
gridsearch = GridSearchCV(model, hyperparameters, cv=5)

# fitting the data to the grid for searching
best_model = gridsearch.fit(X, Y)

# let's see the best hyperparameter values for the best model
print('Best Model Hyperparameters: \n', best_model.best_estimator_.get_params(), '\n')
# mean accuracy for the best model
print('Best Model accuracy (train data): ', best_model.score(X, Y))
print('Best Model accuracy (test data): ', best_model.score(Xtest, Ytest))
# Deploy best_model_GS
```

```python
# test confusion matrix
target_names = ['1', '2']
conf_mat = confusion_matrix(Ytest, test_predictions)
f, ax= plt.subplots(1, 1, figsize=(7,6), dpi=100)
sb.heatmap(conf_mat, annot=True, fmt='.0f', ax = ax, annot_kws={"size":18})
ax.set_xticklabels(target_names, fontsize = 15)
ax.set_yticklabels(target_names, fontsize = 15)
ax.set_xlabel('Predicted labels')
ax.set_ylabel('True labels'); 
ax.set_title('Confusion Matrix Logistic Regression (Validation)');
plt.show()

print('+=+'*18)
print('+=++=++=++=+ Classification Report Project Data'+' '+'+=++=+')
print('+=+'*18)
print(classification_report(Ytest, test_predictions, target_names=target_names))
```

**RandomizedSearchCV**

```python
# Let's use use gridsearch to finetune our logistic regression model
model = LogisticRegression()

# create list of candidate penalty hyperparameters
penalty = ['l1','l2']

# create range of candidate regularization hyperparameters
C = np.logspace(0,4,10)

# create a dictionary hyperparameter candidates
hyperparameters = dict(C=C, penalty=penalty)

# create a grid search object and pass in the hyperparameters defined
randomzedsearch = RandomizedSearchCV(model, hyperparameters, cv=5, n_iter=100, random_state=1)

# fitting the data to the grid for searching
best_model = randomzedsearch.fit(X, Y)

# let's see the best hyperparameter values for the best model
print('Best Model Hyperparameters: \n', best_model.best_estimator_.get_params(), '\n')
# mean accuracy for the best model
print('Best Model accuracy (train data): ', best_model.score(X, Y))
print('Best Model accuracy (test data): ', best_model.score(Xtest, Ytest))
```
